## Hierarchical Clustering
Here we show the iris dataset clustered using hierarchical clustering.  Note that we don't have to specify the number of clusters beforehand, we can generate a dendogram and then make decisions about the best cluster from the dendogram.  We want to explore the output of different linkages for hierarchical clustering, so we'll create a simple function that shows the exact same thing except for a linkage parameter.

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris

%matplotlib inline

iris = load_iris()

def hclus_linkage(data, ltype):
  plt.figure(figsize=(10,10))

  linkage_matrix = linkage(data, ltype)
  dendrogram(
      linkage_matrix,
      truncate_mode="lastp",  # show only the last p merged clusters
      p=40,  # show only the last p merged clusters
      show_leaf_counts=True,  # numbers in brackets are counts, others idx
      leaf_rotation=60.,
      leaf_font_size=8.,
      show_contracted=True,  # to get a distribution impression in truncated branches
  )
  plt.title('Hierarchical Clustering Dendrogram')
  plt.show()
```

## Ward Linkage

Here we use *ward* linkage, something which I don't know much about, but seems to create a very nice looking and believable dendogram.

```python
hclus_linkage(iris.data, 'ward')
```

## Single Linkage

Here we'll do the same thing, but show how the iris dataset clusters using *single linkage*.

```python
hclus_linkage(iris.data, 'single')
```

## Complete Linkage

Same thing, but with complete linkage.

```python
hclus_linkage(iris.data, 'complete')
```

## Average Linkage

Same thing with average linkage

```python
hclus_linkage(iris.data, 'average')
```
